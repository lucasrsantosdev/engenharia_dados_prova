```python
# Celula 1: importações básicas
import re
from datetime import datetime

print("Bibliotecas importadas com sucesso!")
```

```python
# Celula 2: Função para validar CPF
def validar_cpf(cpf: str) -> bool:
    """Valida CPF no formato XXX.XXX.XXX-XX"""
    if cpf is None:
        return False
    pattern = r'^\d{3}\.\d{3}\.\d{3}-\d{2}$'
    return bool(re.match(pattern, cpf))

# Teste rápido
print(validar_cpf("123.456.789-10"))  # True
print(validar_cpf("12345678910"))     # False
```

```python
# Celula 3: Função para validar CEP
def validar_cep(cep: str) -> bool:
    """Valida CEP no formato XXXXX-XXX"""
    if cep is None:
        return False
    pattern = r'^\d{5}-\d{3}$'
    return bool(re.match(pattern, cep))

# Teste rápido
print(validar_cep("12345-678"))  # True
print(validar_cep("12345678"))   # False
```

```python
# Celula 4: Função para validar email simples
def validar_email(email: str) -> bool:
    """Valida email simples"""
    if email is None:
        return False
    pattern = r'^[^@]+@[^@]+\.[^@]+$'
    return bool(re.match(pattern, email))

# Teste rápido
print(validar_email("teste@email.com"))   # True
print(validar_email("testeemail.com"))    # False
```

```python
# Celula 5: Função para validar data no formato YYYY-MM-DD
def validar_data(data_str: str) -> bool:
    """Valida formato YYYY-MM-DD"""
    try:
        datetime.strptime(data_str, '%Y-%m-%d')
        return True
    except:
        return False

# Teste rápido
print(validar_data("2026-03-18"))  # True
print(validar_data("18/03/2026"))  # False
```

```python
# Celula 6: Função para validar status
def validar_status(status: str) -> bool:
    """Status só pode ser ativo, inativo ou suspenso"""
    return status in ["ativo", "inativo", "suspenso"]

# Teste rápido
print(validar_status("ativo"))     # True
print(validar_status("pendente"))  # False
```

```python
# Celula 7: Função para verificar campos obrigatórios
def campos_obrigatorios(dados: dict, campos: list) -> bool:
    """Verifica se campos obrigatórios não são nulos ou vazios"""
    for campo in campos:
        if campo not in dados or dados[campo] is None or str(dados[campo]).strip() == "":
            return False
    return True

# Teste rápido
dados = {"nome": "Lucas", "cpf": "123.456.789-10"}
print(campos_obrigatorios(dados, ["nome", "cpf"]))  # True
print(campos_obrigatorios(dados, ["nome", "email"]))  # False
```
